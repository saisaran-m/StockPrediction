# Stock Price Prediction using Supervised Machine Learning

## 📌 Project Overview & Objectives
This project implements an end-to-end **Supervised Machine Learning** pipeline to analyze historical stock market data, engineer technical indicators, train regression and classification models, evaluate performance out-of-sample, and run financial trading strategy backtests.

### Key Steps Covered:
1. **Data Ingestion**: Fetching historical stock price & volume data via `yfinance`.
2. **Exploratory Data Analysis (EDA)**: Visualizing price trends, moving averages, daily returns, and distributions.
3. **Feature Engineering**: Creating technical indicators (SMA, EMA, RSI, MACD, Bollinger Bands, Volatility, Price Lags).
4. **Supervised Model Training**: Training Linear Regression, Ridge, Random Forest, XGBoost, and Support Vector Machines.
5. **Out-of-Sample Evaluation**: Measuring MAE, RMSE, R² Score, Accuracy, and F1-Score.
6. **Financial Strategy Backtest**: Simulating strategy returns against a Buy & Hold benchmark.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set plotting aesthetics
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 11

# Import custom src modules
from src.data_loader import fetch_stock_data
from src.features import build_supervised_dataset
from src.models import SupervisedStockPredictor
from src.evaluate import run_trading_backtest
```

## Step 1: Historical Data Ingestion & EDA
We fetch historical daily stock data for Apple Inc. (`AAPL`) over a 2-year horizon.

```python
ticker = 'AAPL'
df = fetch_stock_data(ticker=ticker, period='2y')
print(f"Loaded {len(df)} trading days for {ticker}")
    df.tail()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 6))
ax1.plot(df.index, df['Close'], label='Close Price ($)', color='#06b6d4', linewidth=2)
ax1.set_ylabel('Stock Price ($)', color='#06b6d4')
ax1.set_title(f'{ticker} Historical Stock Price & Trading Volume', fontsize=14, fontweight='bold')

ax2 = ax1.twinx()
ax2.bar(df.index, df['Volume'], label='Volume', color='#6366f1', alpha=0.3)
ax2.set_ylabel('Volume', color='#6366f1')
ax2.grid(False)

plt.show()
```

## Step 2: Technical Feature Engineering & Target Construction
We calculate technical indicators as features for supervised learning:
- **Moving Averages**: SMA 10, 20, 50, 200 and EMA 12, 26
- **Momentum**: Relative Strength Index (RSI 14), MACD & Signal
- **Volatility**: Bollinger Bands & 20-day Rolling Volatility
- **Lags**: $Close_{t-1}, Close_{t-2}, Close_{t-3}, Close_{t-5}$
- **Supervised Target**: Next Day Close Price ($Close_{t+1}$)

```python
X, y, clean_df = build_supervised_dataset(df, target_type='regression')
print(f"Supervised Feature Matrix X shape: {X.shape}")
print(f"Supervised Target Vector y shape: {y.shape}")
X.head()
```

## Step 3: Supervised Model Training & Time-Series Evaluation
We train multiple supervised algorithms (Linear Regression, Ridge, Random Forest, XGBoost, SVR) using chronological time-series train/test splitting.

```python
predictor = SupervisedStockPredictor(target_type='regression', test_size=0.2)
results, summary_df = predictor.train_and_evaluate(X, y)
summary_df
```

```python
# Plot Actual vs Predicted Stock Prices for Best Model
best_model_name = predictor.best_model_name
best_preds = results[best_model_name]['predictions']
y_test = results[best_model_name]['y_test']

plt.figure(figsize=(14, 6))
plt.plot(y_test.index, y_test.values, label='Actual Close Price', color='#06b6d4', linewidth=2)
plt.plot(y_test.index, best_preds, label=f'Predicted ({best_model_name})', color='#10b981', linestyle='--', linewidth=2)
plt.title(f'Out-of-Sample Price Forecast: {best_model_name} on {ticker}', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Stock Price ($)')
plt.legend()
plt.show()
```

## Step 4: Technical Feature Importance
We analyze which technical indicators contributed most significantly to the supervised predictions.

```python
feat_imp = predictor.get_feature_importance(list(X.columns))
if not feat_imp.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=feat_imp.head(10), x='Importance', y='Feature', palette='viridis')
    plt.title('Top 10 Technical Indicators by Feature Importance', fontsize=13, fontweight='bold')
    plt.show()
```

## Step 5: Financial Strategy Backtest & Conclusion
We evaluate trading signals derived from the supervised model forecast against a passive Buy & Hold benchmark.

```python
backtest = run_trading_backtest(clean_df, best_preds, target_type='regression')

print("=== BACKTEST RESULTS ===")
print(f"Supervised Strategy Total Return: {backtest['total_strategy_return']:.2f}%")
print(f"Buy & Hold Benchmark Return:   {backtest['total_benchmark_return']:.2f}%")
print(f"Sharpe Ratio:                   {backtest['sharpe_ratio']:.2f}")
print(f"Max Drawdown:                   {backtest['max_drawdown']:.2f}%")
print(f"Signal Win Rate:                {backtest['win_rate']:.2f}%")

# Plot Equity Curves
dates = pd.to_datetime(backtest['dates'])
plt.figure(figsize=(14, 6))
plt.plot(dates, backtest['strategy_equity'], label='Supervised Strategy Equity', color='#6366f1', linewidth=2.5)
plt.plot(dates, backtest['buy_hold_equity'], label='Buy & Hold Benchmark', color='#9ca3af', linestyle='--', linewidth=1.8)
plt.title('Financial Backtest: Cumulative Growth of $1.00 Investment', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Portfolio Growth')
plt.legend()
plt.show()
```
